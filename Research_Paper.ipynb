```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the charts
categories = ['Surgery', 'Chemotherapy', 'Radiotherapy', 'Targeted Therapies', 'Immunotherapies']
urban_access = [90, 85, 80, 50, 40]  # Approximate access percentages in urban areas
rural_access = [60, 55, 50, 25, 20]  # Approximate access percentages in rural areas
cost_factor = [70, 60, 50, 90, 95]  # Approximate cost impact for each treatment

# Plotting access differences between Urban and Rural Areas
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# Bar chart for Urban vs Rural access
ax[0].barh(categories, urban_access, color='lightblue', label='Urban Access')
ax[0].barh(categories, rural_access, color='lightcoral', label='Rural Access')

ax[0].set_xlabel('Access Percentage (%)')
ax[0].set_title('Cancer Treatment Access: Urban vs Rural')
ax[0].legend()

# Bar chart for Cost Factor of Different Treatments
ax[1].barh(categories, cost_factor, color='orange')
ax[1].set_xlabel('Cost Impact (%)')
ax[1].set_title('Cost Impact on Cancer Treatments')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
subtypes = ['Luminal A', 'Luminal B', 'HER2+', 'TNBC']
percentages = [35, 25, 20, 20]
colors = ['#66c2a5', '#fc8d62', '#8da0cb', '#e78ac3']

# Pie Chart
fig1, ax1 = plt.subplots(figsize=(6, 6))
ax1.pie(percentages, labels=subtypes, autopct='%1.1f%%', startangle=90, colors=colors)
ax1.set_title('Breast Cancer Subtype Distribution in Bangladesh')

# Bar Chart
fig2, ax2 = plt.subplots(figsize=(8, 6))
ax2.barh(subtypes, percentages, color=colors)
ax2.set_xlabel('Percentage (%)')
ax2.set_title('Breast Cancer Subtype Distribution in Bangladesh')
ax2.invert_yaxis()  # Highest value on top

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Chart 1: Stage at Diagnosis - Bangladesh vs High-Income Countries
stages = ['Stage I', 'Stage II', 'Stage III', 'Stage IV']
bangladesh_data = [10, 40, 30, 20]
hic_data = [50, 30, 15, 5]

x = np.arange(len(stages))
width = 0.35

fig1, ax1 = plt.subplots(figsize=(8, 6))
ax1.bar(x - width/2, bangladesh_data, width, label='Bangladesh', color='salmon')
ax1.bar(x + width/2, hic_data, width, label='High-Income Countries', color='lightblue')

ax1.set_ylabel('Percentage of Patients (%)')
ax1.set_title('Stage at Diagnosis: Bangladesh vs High-Income Countries')
ax1.set_xticks(x)
ax1.set_xticklabels(stages)
ax1.legend()

# Chart 2: Contributing Factors to Diagnostic Delay
factors = ['Lack of awareness', 'Limited diagnostic facilities', 'Financial barriers',
           'Cultural stigma', 'System delays']
contributions = [35, 25, 20, 10, 10]

fig2, ax2 = plt.subplots(figsize=(8, 6))
ax2.barh(factors, contributions, color='orange')
ax2.set_xlabel('Estimated Contribution (%)')
ax2.set_title('Contributing Factors to Diagnostic Delay in Bangladesh')
ax2.invert_yaxis()
# Chart 3: Screening Coverage and Stage I Diagnosis
categories = ['Screening Coverage', 'Stage I Diagnosis']
hic_values = [75, 50]
bangladesh_values = [10, 10]

x = np.arange(len(categories))
width = 0.35

fig3, ax3 = plt.subplots(figsize=(8, 6))
ax3.bar(x - width/2, bangladesh_values, width, label='Bangladesh', color='salmon')
ax3.bar(x + width/2, hic_values, width, label='High-Income Countries', color='lightblue')

ax3.set_ylabel('Percentage (%)')
ax3.set_title('Screening Coverage and Early Detection Comparison')
ax3.set_xticks(x)
ax3.set_xticklabels(categories)
ax3.legend()

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Barriers data
labels = ['Public Awareness', 'Health System']
public_awareness = [60, 0]  # Sum of 30+20+10
health_system = [0, 40]     # Sum of 20+20

# Stacking data
x = range(len(labels))

plt.figure(figsize=(8, 6))
plt.bar(x, public_awareness, label='Public Awareness Barriers', color='#fdb462')
plt.bar(x, health_system, bottom=public_awareness, label='Health System Barriers', color='#80b1d3')

plt.xticks(x, labels)
plt.ylabel('Estimated Total Impact (%)')
plt.title('Barriers to Early Breast Cancer Diagnosis in Bangladesh')
plt.legend()
plt.tight_layout()
plt.show()
```
